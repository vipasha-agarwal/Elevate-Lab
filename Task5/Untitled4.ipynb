```python
numeric_df = df.select_dtypes(include=['number'])

print("\nCorrelation Heatmap:")
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("\nScatterplots:")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if col != numeric_cols[0]:  # skip ID or similar non-target columns
        plt.figure(figsize=(6,4))
        sns.scatterplot(x=df[numeric_cols[0]], y=df[col])
        plt.title(f'{numeric_cols[0]} vs {col}')
        plt.show()
```

```python
# Let's generate the PDF report from EDA

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fpdf import FPDF

# Load the data again
df = pd.read_csv('/content/train.csv')

# Start PDF
class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 14)
        self.cell(0, 10, 'Exploratory Data Analysis (EDA) Report', 0, 1, 'C')
        self.ln(10)

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1, 'L')
        self.ln(4)

    def chapter_body(self, text):
        self.set_font('Arial', '', 11)
        self.multi_cell(0, 10, text)
        self.ln()

pdf = PDF()
pdf.add_page()

# Basic Information
pdf.chapter_title('Basic Information:')
basic_info = str(df.info(buf=None))
pdf.chapter_body(basic_info)

# Describe
pdf.chapter_title('Summary Statistics:')
desc = df.describe().to_string()
pdf.chapter_body(desc)

# Missing Values
pdf.chapter_title('Missing Values:')
missing = df.isnull().sum().to_string()
pdf.chapter_body(missing)

# Value Counts for Categorical Columns
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    pdf.chapter_title(f'Value Counts: {col}')
    vc = df[col].value_counts().to_string()
    pdf.chapter_body(vc)

# Correlation Heatmap
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,7))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('/content/heatmap.png')
plt.close()

# Add Heatmap to PDF
pdf.add_page()
pdf.image('/content/heatmap.png', x=10, w=190)

# Correlation histogram
df.hist(figsize=(15,10), bins=30)
plt.suptitle('Feature Distributions')
plt.tight_layout()
plt.savefig('/content/histogram.png')
plt.close()

# Add Histogram to PDF
pdf.add_page()
pdf.image('/content/histogram.png', x=10, w=190)

# Correlation Boxplots
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.savefig('/content/boxplots.png')
    plt.close()

# Add Boxplot to PDF
pdf.add_page()
pdf.image('/content/boxplots.png', x=10, w=190)

# Correlation ScatterPlots
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if col != numeric_cols[0]:  # skip ID or similar non-target columns
        plt.figure(figsize=(6,4))
        sns.scatterplot(x=df[numeric_cols[0]], y=df[col])
        plt.title(f'{numeric_cols[0]} vs {col}')
        plt.tight_layout()
        plt.savefig('/content/scatterPlot.png')
        plt.close()

# Add ScatterPlot to PDF
pdf.add_page()
pdf.image('/content/scatterPlot.png', x=10, w=190)

# Observation and summary
pdf.chapter_title('Summary of Findings')
summary = """
- Most passengers were young adults.
- Gender (not shown here) and class had strong influences on survival.
- Age and Fare showed some positive correlation.
- Few outliers present in Age and Fare columns.
- Dataset is moderately clean but requires handling missing Age values.
"""
pdf.chapter_body(summary)

# Save PDF
pdf.output('/content/EDA_Report.pdf')
```
